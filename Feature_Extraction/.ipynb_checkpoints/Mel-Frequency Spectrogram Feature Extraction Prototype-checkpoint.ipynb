## Outline

Feature extraction for this project includes the following steps:

1. Crop and align all audio files for each of the 6 input classes for each of the 40 participants (Ableton)
    - Some are at 80bpm, some are at 60bpm, and some at 40bpm.
    - We must extract the middle 6/8 = 0.75 seconds from each to ensure that the length of all samples is the same. 
    - If substantial information is lost, we can pad the shorter pieces of data such that the total length is 6/4 = 1.5 seconds long. 
    - Each participants data for a class is exported as a single audio file from this step.
2. Segment the audio files. 
3. Calculate the Mel Frequency Cepstrum Coefficients (MFCC).
    - Adjustable parameter: Number of MFCC coefficients.
4. Save and export as images into file system (one sub-folder per class, `arbitrary_names.png`)

```python
# IMPORT BOX #
import numpy as np
import scipy
import matplotlib.pyplot as plt
import pandas
from scipy.io import wavfile
import scipy.io

import librosa
import librosa.display
import IPython.display as ipd
```

```python
def segment_audio(FS, waves, in_len, out_len=0.75):
    """ 
    Returns a list of 1D numpy arrays of `out_len` snippets of samples from an input audio file.
    
    Keyword arguments:
    
    FS --         Sampling rate.
    waves --      Input audio waveforms. 
    in_len --     Length (in SECONDS) of each gesture in the recording (Sample BPM/60).
    
    Optional arguments: 
    
    out_len --    Desired length (in SECONDS) of the gestures in the output (will be symmetrically trimmed).
                  Must be less than or equal to `in_len`.
                  Defaults to 0.75 seconds.
    """
    num_samples = int(len(waves)/(FS*in_len)) # Assuming one per second.
    # TODO: Add contingency for 1 vs. 2-channel input waveforms. 
    wave = (waves[:,0] + waves[:,1])*0.5 # 
    sample_list = []
    
    pad = (in_len - out_len)/2
    
    for i in range(num_samples):
        wave_start = int(i*FS*in_len+pad*FS)
        wave_end = int((i+1)*FS*in_len-pad*FS)
        sample_list.append(wave[wave_start:wave_end])

    return sample_list
```

```python
#########################
# TESTING THE SEGMENTER #
#########################

file_path = '../Baseline_Models/raw_data/Silence_01.wav'
FS, audio = wavfile.read(file_path)
seg_list = segment_audio(FS, audio, 1, )
```

```python
plt.plot(seg_list[1])
plt.title('Sample from `Silence_01.wav`')
plt.xlabel('Sample Number')
plt.ylabel('Amplitude')
plt.show()
```

## Mel Filter Banks

```python
filter_banks = librosa.filters.mel(n_fft=2048, sr=FS, n_mels=100)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(filter_banks, 
                         sr=FS, 
                         x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()
```

```python
mel_spectrogram = librosa.feature.melspectrogram(seg_list[1], sr=FS, n_fft=2048, hop_length=512, n_mels=100)
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
plt.figure(figsize=(10, 10))
librosa.display.specshow(log_mel_spectrogram, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=FS)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
def save_spec_im(arry, name):
#     plt.imshow(arry, cmap='Greys')
#     plt.savefig('{}.png'.format(name))
    plt.imsave('{}.png'.format(name), arry, cmap='gray')
```

```python
save_spec_im(log_mel_spectrogram, 'silence')
```
